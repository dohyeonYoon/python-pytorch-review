# 1.4 텐서의 덧셈

```python
x= torch.rand(2,2)
y= torch.rand(2,2)
print(x)
print(y)
```

```python
x+y # 두 텐서의 합
```

```python
torch.add(x,y) #두 텐서의 합 다른표현
```

```python
y.add(x) # y에 x를 더한다
```

```python
print("원래 y: ", y)
y.add_(x)
print("y= y+x: ", y)
```

# 1.5 텐서의 크기 변환하기 

```python
x= torch.rand(8,8)
print(x)
```

```python
# x를 1차원으로 변환

a= x.view(64) #크기를 바꿔주는 view 8x8 = 64
print(a.size())
```

```python
# x를 3차원으로 변환

b= x.view(4,4,4) # -1 은 원래 크기가 되게 하는 값 8x8 > -1x4x4 즉, 4x4x4 
print(b.size())
print(b)

# -1은 원래 크기가 되게 하는 값이 자동으로 지정되기 떄문에 한 번만 사용할 수 있다.
# 예를들어 x.view(-1,-1,4)는 불가능  
```
